```python
import pandas as pd
import numpy as np
df = pd.DataFrame(
    np.arange(16).reshape(4, 4),
    index=[10, 20, 30, 40],
    columns=list("ABCD")
)
print(df)
df.loc[10]
df.loc[10:20, "B":"D"]
```
