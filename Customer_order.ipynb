```python
#Store each customer's order details (customer name, product, price, category) as tuples inside a list
#tuples inside list
cust_order=[("John","tshirt",50,"cloth"),("Terry","trouser",55,"cloth"),("Nick","laptop",1000,"electronics"),("Gene","pen",2,"stationary"),("John","trouser",75,"cloth"),("Terry","tshirt",30,"cloth"),("Nick","mouse",10,"electronics"),("Gene","tshirt",10,"cloth"),("Gene","desktop",100,"electronics")]
print(cust_order)
```

```python
#Analyze customer orders
#Use a loop to calculate the total amount each customer spends
total_spent={}
for name, product, price, category in cust_order:
    if name in total_spent:
        total_spent[name]+=price
    else:
        total_spent[name]=price
print(total_spent)
```

```python
#If the total purchase value is above $100, classify the customer as a high-value buyer
class_cust={}
for name,price in total_spent.items():
    if price > 100:
        class_cust[name]="high-value"
    else:
        class_cust[name]="regular-value"
print(class_cust)
```

```python
#If it is between $50 and $100, classify the customer as a moderate buyer
class_cust={}
for name,price in total_spent.items():
    if price > 100:
        class_cust[name]="high-value"
    elif 50 <= price <= 100:
        class_cust[name]="moderate-buyer"
    else:
        class_cust[name]="regular-buyer"
print(class_cust)
```

```python
class_cust={}
for name,price in total_spent.items():
    if price > 100:
        class_cust[name]="high-value"
    elif 50 <= price <= 100:
        class_cust[name]="moderate-buyer"
    else:
        class_cust[name]="low-value buyer"
print(class_cust)
```

```python
#Generate business insights
#Calculate the total revenue per product category and store it in a dictionary

total_rev = {}

for name, product, price, category in cust_order:
    total_rev[category] = total_rev.get(category, 0) + price
print(total_rev)
```

```python
#Extract unique products from all orders using a set
unq_prod = set()

for name, product, price, category in cust_order:
    unq_prod.add(product)
    
print(unq_prod)
```

```python
#Use a list comprehension to find all customers who purchased electronics
cust_elect=[]

for name, product, price, category in cust_order:
    if category == "electronics":
        cust_elect.append(name)
print(set(cust_elect))
#electronics_buyers = [name for name, product, price, category in orders if category == "Electronics"]    
```

```python
#Use set operations to find customers who purchased from multiple categories

cust_categories = {}

for name, product, price, category in cust_order:
    if name not in cust_categories:
        cust_categories[name] = set()
    cust_categories[name].add(category)
print(cust_categories)

mult_cust_categories= {}
for name,categories in cust_categories.items():
    if len(categories) >1:
        mult_cust_categories[name]=categories
print(mult_cust_categories)
```

```python
#Identify common customers who bought both electronics and clothing

comm_cust_categories= {}
for name,categories in cust_categories.items():
    if "electronics" in categories and "cloth" in categories:
        comm_cust_categories[name]=categories
print(comm_cust_categories)
```
