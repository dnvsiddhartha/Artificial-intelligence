```python
#Calculate descriptive statistics like mean, median, and standard deviation for numerical columns such as LotArea, YearBuilt, 1stFlrSF, 2ndFlrSF, and SalePrice
num_cols = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF", "SalePrice"]

stats=df[num_cols].agg(["mean","median","std"])
print(stats)
```

```python
#Convert the YearBuilt and YearRemodAdd columns to datetime format (if not converted)
df["YearBuilt"] = pd.to_datetime(df["YearBuilt"], format="%Y")
df["YearRemodAdd"] = pd.to_datetime(df["YearRemodAdd"], format="%Y")
```

```python
df["YearBuilt"].head()
```

```python
#Extract useful components from the date like the year, month, or day
df["BuiltYear"]=df["YearBuilt"].dt.year
df["BuiltMonth"]=df["YearBuilt"].dt.month
df["BuiltDay"]=df["YearBuilt"].dt.day
df["RmodYear"]=df["YearRemodAdd"].dt.year
df["RmodMonth"]=df["YearRemodAdd"].dt.month
df["RmodDay"]=df["YearRemodAdd"].dt.day
```

```python
#Calculate the time difference between the year the house was built and the year it was remodeled
df["DiffYear"]=df["BuiltYear"]-df["RmodYear"]
# To view multiple columns, use a list of column names, not a tuple
df[["DiffYear", "YearBuilt", "YearRemodAdd"]].head()  # Changed tuple to list with double brackets
```

```python
#Count the number of occurrences of each category in categorical features
cat_cols=["Neighborhood","BldgType","HouseStyle"]
#cat_cols=df.select_dtypes(include="object").columns

#for col in cat_cols:
#    print(f"\ncolumn: {col}")
#    print(df[col].value_counts())
    
value_counts_dict = {}
for col in cat_cols:
    value_counts_dict[col] = df[col].value_counts()

print(value_counts_dict)
```
